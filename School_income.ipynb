```python
import pandas as pd 
import requests 
from io import StringIO
import seaborn as sns 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
text= open('secret.txt', 'r')
url  = str(text.readlines()[0])
```

```python
url = 'https://api.apispreadsheets.com/data/obgmw0lRjk2CWKpM/'
res = requests.get(url).json()
```

```python
res = res['data'][17:]
```

```python
res[0]
```

```python
data = pd.DataFrame(columns=res[0].values())
```

```python
list(res[1].values())
```

```python
for x in range(1, len(res)):
    tuition = np.around((list(res[x].values())[1] *5) /3, decimals=2)
    sales = np.around((list(res[x].values())[2] *5) /3, decimals=2)
    data= data.append({
        'Date': list(res[x].values())[0],
        'Tuition ': tuition,
        'Services / Sales': sales,
        'Monthly Grand Total ': tuition + sales

    },ignore_index=True)
```

```python
data
```

```python
plt.style.use('seaborn')
fig = plt.figure(figsize=(15,8))
ax = sns.barplot('Date', 'Tuition ',data=data, color='dodgerblue')

plt.title('2023 Tuition', color='white',weight='bold', fontsize=30)
plt.xlabel(' ')
plt.ylabel('',fontsize=16, color='white')
ax.set_facecolor('dimgrey')
fig.set_facecolor('dimgrey')
plt.xticks(rotation=45,ha='right', color='white', fontsize=15)
plt.yticks( color='white', fontsize=15)
ax.yaxis.set_major_formatter('${x:,.2f}')
plt.ylim(50000,350000)



for i , g in enumerate(data['Tuition ']):
    if g != 0:
        if i ==2:
            plt.text(i, g+2000, '${:,.2f}'.format(g), fontsize=20.7, color='dodgerblue', weight='bold',horizontalalignment='center')
            plt.text(i, g+2000, '${:,.2f}'.format(g), fontsize=20, color='white', weight='bold',horizontalalignment='center')
            
        else:
            plt.text(i, g+13000, '${:,.2f}'.format(g), fontsize=20.7, color='dodgerblue', weight='bold',horizontalalignment='center')
            plt.text(i, g+13000, '${:,.2f}'.format(g), fontsize=20, color='white', weight='bold',horizontalalignment='center',)
        
plt.show()
```

```python
plt.style.use('seaborn')
fig = plt.figure(figsize=(15,8))
ax = sns.barplot('Date', 'Services / Sales',data=data, color='lightcoral')
plt.title('2023 Sales', color='white',weight='bold', fontsize=30)
plt.xlabel(' ')
plt.ylabel('',fontsize=16, color='white')
ax.set_facecolor('dimgrey')
fig.set_facecolor('dimgrey')
plt.xticks(rotation=45,ha='right', color='white', fontsize=15)
plt.yticks( color='white', fontsize=15)
ax.yaxis.set_major_formatter('${x:,.2f}')
plt.ylim(5000,40000)



for i , g in enumerate(data['Services / Sales']):
    if g != 0:
        
        if i ==4:
            plt.text(i, g+400, '${:,.2f}'.format(g), fontsize=20.7, color='lightcoral', weight='bold',horizontalalignment='center')
            plt.text(i, g+400, '${:,.2f}'.format(g), fontsize=20, color='white', weight='bold',horizontalalignment='center')
            
        else:
            plt.text(i, g+1700, '${:,.2f}'.format(g), fontsize=20.7, color='lightcoral', weight='bold',horizontalalignment='center')
            plt.text(i, g+1700, '${:,.2f}'.format(g), fontsize=20, color='white', weight='bold',horizontalalignment='center',)
        
plt.show()
```
